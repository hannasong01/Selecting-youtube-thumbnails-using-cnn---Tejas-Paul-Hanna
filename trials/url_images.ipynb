```python
from google.colab import drive
from pytube import YouTube
import os
import subprocess
import pandas as pd
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
import shutil
```

```python

def download_video_thumbnail(video_url, output_folder, percentile=50):
    try:
        # Create a YouTube object with the video URL
        yt = YouTube(video_url)

        # Get the video length in seconds
        video_length = yt.length

        # Calculate the time point based on the given percentile
        time_point = int((percentile / 100) * video_length)

        # Use ffmpeg to extract a frame at the specified time point
        output_file_path = os.path.join(output_folder, f"{yt.video_id}_{percentile}th_percentile.jpg")
        subprocess.run(["ffmpeg", "-i", yt.streams.get_highest_resolution().url, "-ss", str(time_point), "-vframes", "1", output_file_path])

        print(f"Thumbnail for video ID {yt.video_id} at {percentile}th percentile saved to {output_file_path}")

        return output_file_path

    except Exception as e:
        print(f"An error occurred for video URL {video_url}: {e}")
        return None

def save_thumbnail_to_drive(thumbnail_path, drive_folder_path):
    try:
        # Create the 'thumbnails' folder in Google Drive if it doesn't exist
        os.makedirs(drive_folder_path, exist_ok=True)

        # Copy the thumbnail image to Google Drive using shutil
        shutil.copy(thumbnail_path, drive_folder_path)

        print(f"Thumbnail copied to Google Drive folder.")
    except FileNotFoundError:
        print(f"Error: File not found - {thumbnail_path}")
    except Exception as e:
        print(f"An error occurred during file copy: {e}")

def process_multiple_urls_from_drive(dataset_path, output_folder, drive_folder_path, percentile=50):
    # Mount Google Drive
    drive.mount('/content/drive')

    # Read the dataset from Google Drive
    df = pd.read_csv(dataset_path)

    # Extract video URLs from the dataset
    video_urls = df['video_url']

    # Iterate over the list of video URLs
    for video_url in video_urls:
        thumbnail_path = download_video_thumbnail(video_url, output_folder, percentile)
        if thumbnail_path:
            save_thumbnail_to_drive(thumbnail_path, drive_folder_path)

# Example usage:
dataset_path = '/content/drive/MyDrive/Neural Network And Deep Learning/final_yt_data.csv'  # Adjust the path to your dataset
output_folder = "thumbnails"
drive_folder_path = '/content/drive/MyDrive/Neural Network And Deep Learning/thumb'  # Adjust the path to your Google Drive folder
percentile = 30  # You can adjust the percentile as needed

process_multiple_urls_from_drive(dataset_path, output_folder, drive_folder_path, percentile)
```

```python
def download_video_thumbnail(video_url, output_folder, drive_folder, percentile=50):
    try:
        # Create a YouTube object with the video URL
        yt = YouTube(video_url)

        # Get the video length in seconds
        video_length = yt.length

        # Calculate the time point based on the given percentile
        time_point = int((percentile / 100) * video_length)

        # Use ffmpeg to extract a frame at the specified time point
        output_file_path = os.path.join(output_folder, f"{yt.video_id}_{percentile}th_percentile.jpg")
        subprocess.run(["ffmpeg", "-i", yt.streams.get_highest_resolution().url, "-ss", str(time_point), "-vframes", "1", output_file_path])

        print(f"Thumbnail for video ID {yt.video_id} at {percentile}th percentile saved to {output_file_path}")

        return output_file_path

    except Exception as e:
        print(f"An error occurred for video URL {video_url}: {e}")
        return None

def save_thumbnail_to_drive(thumbnail_path, drive_folder_id):
    gauth = GoogleAuth()
    gauth.LocalWebserverAuth()
    drive = GoogleDrive(gauth)

    # Upload the thumbnail image to Google Drive
    drive_thumbnail = drive.CreateFile({'title': os.path.basename(thumbnail_path), 'parents': [{'id': drive_folder_id}]})
    drive_thumbnail.Upload()

    print(f"Thumbnail uploaded to Google Drive folder.")

def process_multiple_urls_from_drive(dataset_path, output_folder, drive_folder_id, percentile=50):
    # Mount Google Drive
    drive.mount('/content/drive')

    # Read the dataset from Google Drive
    df = pd.read_csv(dataset_path)

    # Extract video URLs from the dataset
    video_urls = df['video_url']

    # Iterate over the list of video URLs
    for video_url in video_urls:
        thumbnail_path = download_video_thumbnail(video_url, output_folder, drive_folder_id, percentile)
        if thumbnail_path:
            save_thumbnail_to_drive(thumbnail_path, drive_folder_id)

# Example usage:
dataset_path = '/content/drive/MyDrive/Neural Network And Deep Learning/final_yt_data.csv'  # Adjust the path to your dataset
output_folder = "thumbnails"
drive_folder_id = 'Neural Network And Deep Learning'  # Replace with your Google Drive folder ID
percentile = 50  # You can adjust the percentile as needed

process_multiple_urls_from_drive(dataset_path, output_folder, drive_folder_id, percentile)
```
