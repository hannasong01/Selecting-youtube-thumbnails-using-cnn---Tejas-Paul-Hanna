```python
import csv
import json
from googleapiclient.discovery import build
import pandas as pd
import numpy as np
```

```python
def initialize_youtube_api(api_key):
    return build('youtube', 'v3', developerKey=api_key)
```

```python
def get_channel_name(youtube, channel_id):
    request = youtube.channels().list(
        part="snippet",
        id=channel_id
    )
    response = request.execute()

    return response['items'][0]['snippet']['title']
```

```python
def get_videos_from_channel(youtube, channel_id, max_results=50):
    channel_name = get_channel_name(youtube, channel_id)

    request = youtube.search().list(
        part="snippet",
        channelId=channel_id,
        type="video",
        order="date",
        maxResults=max_results
    )
    response = request.execute()

    videos = []

    for item in response['items']:
        video_id = item['id']['videoId']
        video_info = get_video_info(youtube, video_id, channel_name)
        videos.append(video_info)

    return videos
```

```python
def get_video_info(youtube, video_id, channel_name):
    request = youtube.videos().list(
        part="snippet,statistics",
        id=video_id
    )
    response = request.execute()

    video_info = {
        "channel_name": channel_name,
        "published_date": response['items'][0]['snippet']['publishedAt'],
        "view_count": response['items'][0]['statistics']['viewCount'],
        #"like_count": response['items'][0]['statistics']['likeCount'],
        "video_url": f"https://www.youtube.com/watch?v={video_id}",
        "title": response['items'][0]['snippet']['title']
    }

    return video_info
```

```python
def scrape_videos(api_key, channel_ids, max_results=50):
    youtube = initialize_youtube_api(api_key)

    all_videos = []

    for channel_id in channel_ids:
        videos = get_videos_from_channel(youtube, channel_id, max_results)
        all_videos.extend(videos)

    return all_videos
```
